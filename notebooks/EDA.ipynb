```python
df_long = df.melt(id_vars=['Name', 'Year', 'Platform', 'Platform_Generation', 'Platform_Company','Global_Sales'],
                  var_name = 'Country',
                  value_name='Sales')

df_long['Country'] = df_long['Country'].apply(lambda x: x.split('_')[0])
df_long.head()
```
