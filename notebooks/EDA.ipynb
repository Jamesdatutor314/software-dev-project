```python
"""Exploratory data analysis of video game sales."""
# import main libraries
import pandas as pd
import plotly.express as px
```

-----

Featuring video games popular in North America, Japan, Europe, and beyond, 

this dataset only includes titles with a minimum of 100k in global sales.



------

```python
# upload the dataset
URL = "../vgsales.csv"
df = pd.read_csv(URL)
df.sample(10)
```

```python
# basic info
df.info()
```

```python
# summary
df.describe(include="all")
```

```python
df_long = df.melt(
    id_vars=[
        "Name",
        "Year",
        "Platform",
        "Platform_Generation",
        "Platform_Company",
        "Global_Sales",
    ],
    var_name="Country",
    value_name="Sales",
)

df_long["Country"] = df_long["Country"].apply(lambda x: x.split("_")[0])
df_long.head()
```

```python
fig = px.bar(
    df_long,
    x="Year",
    y="Sales",
    color="Country",
    hover_name="Sales",
    barmode="group",
    title="Video Game Sales Each Year Per Country",
    labels={"Sales": "Sales (millions)"},
)
fig.show()
```

```python
fig = px.scatter(df, x="EU_Sales", y="Global_Sales", color="Platform_Company")


fig.show()
```

```python
fig = px.histogram(df, x="Global_Sales")


fig.show()
```
